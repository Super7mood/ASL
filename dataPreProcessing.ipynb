```python
import os
import mediapipe as mp
import cv2
import matplotlib.pyplot as plt
import pickle
```

```python
#mediapipe code from https://mediapipe.readthedocs.io/en/latest/solutions/hands.html
#the code is inspired by https://www.youtube.com/watch?v=n_eMARPqBZI


mpHands = mp.solutions.hands
mpDrawing = mp.solutions.drawing_utils
mpDrawingStyles = mp.solutions.drawing_styles

hands = mpHands.Hands(static_image_mode=True, min_detection_confidence=0.3, max_num_hands=1)
hands
```

```python
dataDirectory = './data'  # Define the path to the data directory

instances = {chr(letter) : 0 for letter in range(ord('A'), ord('Z') + 1)}
processed = {chr(letter) : 0 for letter in range(ord('A'), ord('Z') + 1)}




showTwoImg = 0  # Initialize a counter to track the number of images displayed
counter = 0
noLandmark = []
doubleLandmark = []

data = []
labels = []
sizes = []

# Iterate through each directory in the specified data directory


for directory in os.listdir(dataDirectory):

    print("Processing class " + directory + " images...")

    for imgPath in os.listdir(os.path.join(dataDirectory, directory)):     

        dataAUX = []
        
        counter += 1
        instances[imgPath[0]] += 1
        
        img = cv2.imread(os.path.join(dataDirectory, directory, imgPath))  # Read the image from the file
        
        # Check if the image was loaded successfully
        imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert the image from BGR to RGB color space
        
        results = hands.process(imgRGB)
        
        if not results.multi_hand_landmarks:
            #print(directory + " " + imgPath)
            noLandmark.append(directory + " " + imgPath)
            continue
            
        for handLandmarks in results.multi_hand_landmarks:
            for i in range(len(handLandmarks.landmark)):
                x = handLandmarks.landmark[i].x
                y = handLandmarks.landmark[i].y
                dataAUX.append(x)
                dataAUX.append(y)
                
        if(len(dataAUX)) != 42:
            doubleLandmark.append(directory + " " + imgPath)
            print("Double " + imgPath)
            continue


            
        #     for handLandmarks in results.multi_hand_landmarks:
        #         mpDrawing.draw_landmarks(
        #             imgRGB, #image to draw
        #             handLandmarks, # model output
        #             mpHands.HAND_CONNECTIONS,
        #             mpDrawingStyles.get_default_hand_landmarks_style(),
        #             mpDrawingStyles.get_default_hand_connections_style())
        #     plt.figure()
        #     plt.imshow(imgRGB)
        #     plt.axis('off')

        processed[directory] += 1
        data.append(dataAUX)
        labels.append(directory)
        sizes.append(len(dataAUX))

# plt.show()          
print(noLandmark)
print(len(noLandmark))
print(counter)
print(counter - len(noLandmark))

f = open('data.pickle', 'wb')
pickle.dump({'data': data, 'labels': labels}, f)
f.close()
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python



data = np.asarray(data)
labels = np.asarray(labels)

x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size = 0.3, stratify=labels, shuffle=True)
```

```python
model = RandomForestClassifier()



model.fit(x_train, y_train)



y_predict = model.predict(x_test)

score = accuracy_score(y_predict, y_test)
score
```

```python

mpHands = mp.solutions.hands
mpDrawing = mp.solutions.drawing_utils
mpDrawingStyles = mp.solutions.drawing_styles

cap = cv2.VideoCapture(0)

hands = mpHands.Hands(static_image_mode=True, model_complexity=2 ,min_detection_confidence=0.3, max_num_hands=1)

if not cap.isOpened():
    print("Error: Unable to access the camera.")

else:
    while True:
        xPoints = []
        yPoints = []
        ret, frame = cap.read()
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        dataAUX = []

        H, W, _ = frame.shape

        results = hands.process(frame_rgb)
        if results.multi_hand_landmarks:
            for handLandmarks in results.multi_hand_landmarks:
                mpDrawing.draw_landmarks(
                frame, #image to draw
                handLandmarks, # model output
                mpHands.HAND_CONNECTIONS,
                mpDrawingStyles.get_default_hand_landmarks_style(),
                mpDrawingStyles.get_default_hand_connections_style())
        
            for handLandmarks in results.multi_hand_landmarks:
                dataAUX = []
                for i in range(len(handLandmarks.landmark)):
                    x = handLandmarks.landmark[i].x
                    y = handLandmarks.landmark[i].y
                    dataAUX.append(x)
                    dataAUX.append(y)
                    xPoints.append(x)
                    yPoints.append(y)
            
            x1 = int(min(xPoints) * W)
            y1 = int(min(yPoints) * H)

            x2 = int(max(xPoints) * W)
            y2 = int(max(yPoints) * H)
            
            
            print(len(dataAUX))
            prediction = model.predict([np.asarray(dataAUX)])
            print(prediction[0])

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 0), 4)
            cv2.putText(frame, prediction[0], (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0,0,0), 3, cv2.LINE_AA)
            
        cv2.imshow('frame', frame)
        cv2.waitKey(1) 

cap.release()
cap.destroyALLWindows()
```
